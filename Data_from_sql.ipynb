```python
import sqlite3
import pandas as pd
```

```python
# Connect to local database
conn = sqlite3.connect('/Users/ernestofadel/Desktop/DC_local/Data_files/Database_lite/chembl_36.db')
```

```python
# Your query as SQL
query = """
SELECT
    asy.chembl_id as assay_id,
    asy.description,
    dc.chembl_id as doc_id,
    dc.year,
    asy.confidence_score,
    td.chembl_id as targed_id,
    td.pref_name,
    comp_seq.sequence AS protein_sequence,
    comp_seq.accession AS uniprot_accession,
    asy.variant_id,
    md.chembl_id as molecule_id,
    cs.canonical_smiles,
    a.standard_type,
    a.standard_value,
    a.standard_units,
    a.pchembl_value,
    asy.assay_type,
    asy.assay_organism,
    asy.assay_category,
    asy.assay_tax_id, 
    asy.assay_strain, 
    asy.assay_tissue, 
    asy.assay_cell_type,
    asy.assay_subcellular_fraction, 
    asy.bao_format
FROM activities a
LEFT JOIN assays asy ON a.assay_id = asy.assay_id
LEFT JOIN target_dictionary td ON asy.tid = td.tid
LEFT JOIN docs dc ON a.doc_id = dc.doc_id
LEFT JOIN compound_structures cs ON a.molregno = cs.molregno
LEFT JOIN target_components tc ON td.tid = tc.tid
LEFT JOIN component_sequences comp_seq ON tc.component_id = comp_seq.component_id
LEFT JOIN molecule_dictionary md ON a.molregno = md.molregno
WHERE 
    a.standard_type = 'IC50'
    AND a.standard_units = 'nM'
    AND a.standard_relation = '='
    AND a.data_validity_comment IS NULL
    AND td.target_type = 'SINGLE PROTEIN'
"""

# Execute and load into DataFrame
df = pd.read_sql_query(query, conn)
#conn.close()

print(f"Retrieved {len(df)} records")
df
```
